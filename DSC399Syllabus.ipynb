## Course Description
This course examines advanced applications of neural networks within fields such as natural language processing, computer vision, reinforcement learning, and generative modeling. Students will explore modern neural network architectures, including attention mechanisms, transformers, graph neural networks, and deep generative models. This course will also consider the interpretability and scalability of these modern architectures, along with the ethical considerations of gathering data to train these models.

## Course Purpose
This course serves as a follow-up for DSC 340 and will cover applications of neural networks which are important in the modern world. These applications will also blend machine learning and artificial intelligence, making them increasingly important in modern computing landscape.

## Prerequisities

DSC 340 or permission of the instructor.

## COURSE LEARNING OUTCOMES:
* Understand and implement modern neural network architectures with application to text analysis, computer vision, graph analysis, and generative modeling.
* Evaluate models in terms of interpretability and scalability.
* Discuss real-world applications of the modern neural network architectures along with ethical considerations that come from these applications

### Quizzes
Four quizzes will be given throughout the semester, testing your knowledge of the concepts and algorithms taught in lecture. Students will be given 30 minutes to complete each quiz, which will contain 5-10 short answer questions. A practice quiz will be provided to students at least one week prior to each quiz.

### Homework
Four homeworks will be given throughout the course, each one furthering your knowledge of the algorithms, code implementations, and concepts. For questions that require a coding implemenation, correct and documented code must be provided for full credit. For questions which require a text explaination, the explainations must be provided in the homework notebook in formatted markdown cells. Expect to spend approximately 15 - 20 hours on each homework assignment over the course of a two week period. Note that these assignments will have long run times.

### Attendance and Participation
As this is a flipped classroom, your attendance and participation in all class meetings is required. For each class meeting you will receive an attendance score out of 5, with points being deducted for arriving late and no points received for an
unexcused absence. An excused absence requires a note from the health center or another health care provider, a note from a coach in the case of an athletic absence, or a note from a professor in the case of a field trip. In the case of a family
emergency or other situation that will require an absence, clear communication prior to the absence is required. Any unexcused absences will receive an attendance score of 0. For each class meeting you will also receive a participation score out of 5, which full points being awarded for steadily working on the week’s projects. Points will be deducted for phone use, off topic conversations, etc.

### Final Project
The final project will allow you to apply concepts from the course to a problem that is unique to your interest. The final project must use machine learning to solve a problem, with some restrictions. If you plan to use an algorithm that was covered either in this course or in DSC 340, then you must apply the algorithm to a type of data which we have not covered (for example applying a recurrent neural network to analyze sound data). If you plan to use a type of data that we have used in this course or in DSC 340, then you must use an algorithm that was not covered in either course. Each student must give a 15 minute presentation on their work during the final exam period for the course. More details of the final project can be found on D2L.

### ACADEMIC INTEGRITY:
Academic Integrity is at the heart of the mission and values of the University and is an expectation of all students. Maintaining academic integrity is a reflection of character and personal integrity and is a means to achieving the outcomes of this course and that grades accurately reflect learning and understanding of the course material. Mount Union expects all students to be responsible members of the academic community, so all work submitted in this course should be yours and yours alone. Please refer to the Academic Honesty section in the current Undergraduate Catalog and Student Handbook.

All work you submit with your name on it is expected to be original work. You can consult any outside source, including the internet and AI chats, for help on the class projects, but you are not allowed to copy any solutions you find there
directly. Additionally, you should be able to thoroughly explain how you arrived at your answer for all work you turn in; all assignments are subject to possible verbal discussion which can contribute to your grade. If you work closely with other
classmates on an assignment, please indicate that the solution results from collaboration and list the names of all students who contributed (this is allowed and encouraged). If it can be proven that you copied directly from Chegg, ChatGTP, or
another person on your work (i.e., you are copying your solutions directly from these sources or others), you will receive a zero for the assignment and be reported for academic dishonesty.
